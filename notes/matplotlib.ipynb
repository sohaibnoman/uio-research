```python
from matplotlib import pyplot as plt
```

```python
# default graph

x = [1,2,3]
y = [5,7,4]

x2 = [1,2,3]
y2 = [10,14,12]

plt.plot(x,y, label="Seconf line")
plt.plot(x2,y2, label="First line")

plt.xlabel('Plot number')
plt.ylabel('Important var')

plt.title("The Siiick grapph")

# only needed if more plots are in one plt
plt.legend()
plt.show()
```

```python
# barchart

x = [2,4,6,8,10]
y = [4,6,7,8,2]

x2 = [12,2,4,4,14]
y2 = [7,4,4,6,6]

plt.bar(x, y, label="Barzz", color='blue')
plt.bar(x2, y2, label="Barzz 2", color='#564343')

plt.xlabel('Plot number')
plt.ylabel('Important var')

plt.title("The Siiick grapph")

# only needed if more plots are in one plt
plt.legend()
plt.show()
```

```python
# histogram

population = [22,23,55,45,22,45,67,56,34,45,33,67]

#ids = [x for x in range(len(population))]

#plt.bar(ids, population, label="Pops")
bins = [20,30,40,50,60,70]

plt.hist(population, range(20,70,10), histtype='bar', rwidth=0.7, label="Pooz")

plt.xlabel('Plot number')
plt.ylabel('Important var')

plt.title("The Siiick grapph")

# only needed if more plots are in one plt
plt.legend()
plt.show()
```

```python
# scatterplot (useful for correlation or relation between teo things)

x = range(0,8,1)
y = [4,6,8,3,8,9,4,2]

# s = size
plt.scatter(x,y, label="scit scat", color='k', marker='+', s=100)

plt.xlabel('x')
plt.ylabel('y')

plt.title("The Siiick grapph")

# only needed if more plots are in one plt
plt.legend()
plt.show()
```

```python
# stack plots
# with polygons like here you cant have label

days = [1,2,3,4]

sleeping = [7,8,6,11]
eating = [2,4,6,2]
working = [7,8,5,2]
playing = [8,5,7,10]

# workaround for labels have empty plots
plt.plot([],[],color='m', label="sleeping", linewidth=5)
plt.plot([],[],color='c', label="eating")
plt.plot([],[],color='r', label="working")
plt.plot([],[],color='b', label="playing")

# first is x axes and the rest is y
plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','b'])

plt.xlabel('x')
plt.ylabel('y')

plt.title("The Siiick grapph")

# only needed if more plots are in one plt
plt.legend()
plt.show()
```
